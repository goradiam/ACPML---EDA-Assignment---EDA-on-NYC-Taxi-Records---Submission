## **3 Finding the Optimal Number of Clusters**

Previously, we arbitrarily assumed a number of clusters. Now, however, we will estimate what the correct number of clusters will be by using two techniques:

*   By calculating the **silhouette score**
*   By creating an **elbow plot**

---

The silhouette score is a metric used to evaluate the quality of a clustering solution. It measures how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The score ranges from -1 to +1:
* A score close to +1 indicates that the data points are well-clustered, meaning they are close to other points in the same cluster and far from points in other clusters.
* A score close to 0 indicates that the data points are on or very near the boundary between two clusters.
* A score close to -1 suggests that the data points might have been assigned to the wrong cluster, as they are closer to points in other clusters than to points in their own cluster.

---

An elbow plot is a graphical tool used to determine the optimal number of clusters for clustering algorithms, particularly for k-means clustering. The elbow plot helps in deciding the value of K (the number of clusters) by showing how the within-cluster sum of squares (WSS), or inertia, changes as the number of clusters increases.
